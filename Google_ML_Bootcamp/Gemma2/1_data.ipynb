```python
import zipfile
import os

# 압축을 풀 zip 파일 경로
zip_file_path = 'TL_out.zip'

# 압축을 풀 디렉토리 설정 (현재 디렉토리로 설정)
output_dir = 't/'

# zip 파일 열기
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # 파일 목록 확인 (옵션)
    print(zip_ref.namelist())  # zip 파일에 있는 파일 리스트 출력

    # 압축 풀기
    zip_ref.extractall(output_dir)

print("압축 해제가 완료되었습니다.")
```

```python
import json
import os

# JSON 파일들이 있는 디렉토리 경로
json_dir = 't/'

# 병합된 데이터를 저장할 리스트
data_t = []

# JSON 파일들 하나씩 열기
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        file_path = os.path.join(json_dir, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)  # JSON 파일 읽기
            data_t.append(data)  # 데이터 병합

# 병합된 데이터를 하나의 파일로 저장
with open('data_t.json', 'w', encoding='utf-8') as f:
    json.dump(data_t, f, ensure_ascii=False, indent=4)

print(f"총 {len(data_t)}개의 JSON 파일이 병합되었습니다.")
```

```python
import pandas as pd

# JSON 데이터를 불러온 후 DataFrame으로 변환했다고 가정
# df['list']는 중첩된 리스트를 포함하는 컬럼

# type과 text 값을 저장할 리스트
text_t = []

# df['list']의 각 항목을 순회
for idx, row in data_t.iterrows():
    if isinstance(row['list'], list):  # 'list'가 리스트인지 확인
        for sublist in row['list']:
            if 'list' in sublist:  # 중첩된 'list' 확인
                for item in sublist['list']:
                    if 'audio' in item:  # audio 리스트 확인
                        for audio_item in item['audio']:
                            # 'type'과 'text' 추출
                            text_t.append({
                                'type': audio_item.get('type'),
                                'text': audio_item.get('text')
                            })

# 추출한 데이터를 DataFrame으로 변환
text_t = pd.DataFrame(text_t)

# 결과 확인
print(text_t)
```
