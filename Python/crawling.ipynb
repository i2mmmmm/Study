```python
import requests
from bs4 import BeautifulSoup

# 크롤링할 URL을 입력합니다. 예시로 'https://example.com'을 사용합니다.
url = 'https://eduon.com/itembank/itemlist/19/831/'

# 웹 페이지를 가져옵니다.
response = requests.get(url)

# 웹 페이지가 성공적으로 로드되었는지 확인합니다.
if response.status_code == 200:
    # 웹 페이지의 HTML 내용을 BeautifulSoup을 사용하여 파싱합니다.
    soup = BeautifulSoup(response.text, 'html.parser')

    # 파싱한 HTML에서 제목(title)을 출력해봅니다.
    title = soup.title.string
    print('Page Title:', title)
else:
    print('웹 페이지를 가져오는 데 실패했습니다.')
```

```python
# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(url, 'html.parser')

# 문제의 지문 추출
question_text = soup.select_one('.viewbox').get_text(strip=True)

# 보기 추출
answers = []
answer_rows = soup.select('.table_answer tr')
for row in answer_rows:
    cols = row.find_all('td')
    if len(cols) >= 2:  # 보기 항목이 두 개 이상일 경우
        answers.append(cols[1].get_text(strip=True))

# 결과 출력
print("지문:")
print(question_text)

print("\n보기:")
for idx, answer in enumerate(answers, 1):
    print(f"{idx}. {answer}")
```

```python
import requests
from bs4 import BeautifulSoup

# 크롤링할 URL
url = 'https://eduon.com/itembank/itemlist/19/831/'

# 웹 페이지 요청
response = requests.get(url)

# HTTP 요청이 성공했는지 확인 (status_code 200이면 성공)
if response.status_code == 200:
    print("웹 페이지를 성공적으로 불러왔습니다.")
    html_content = response.text  # HTML 콘텐츠 가져오기

    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(html_content, 'html.parser')

    # 문제 지문 추출
    # 페이지에서 지문을 포함하는 정확한 클래스를 사용해야 함
    # 예시로 .viewbox를 사용했지만 실제 페이지 구조를 확인하여 적절한 선택자 사용해야 함
    question_text = soup.select_one('.viewbox')  # .viewbox 클래스를 가진 첫 번째 요소 선택
    if question_text:
        question_text = question_text.get_text(separator=" ", strip=True)
    else:
        question_text = "지문을 찾을 수 없습니다."

    # 결과 출력
    print("\n지문:")
    print(question_text)

else:
    print(f"웹 페이지를 불러오는 데 실패했습니다. 상태 코드: {response.status_code}")
```
