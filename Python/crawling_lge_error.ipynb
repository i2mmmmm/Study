```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
urls = copied_data.strip().splitlines()

results = []

for idx, url in enumerate(urls):
  try:
    response = requests.get(url, timeout=5)

    soup = BeautifulSoup(response.text, 'html.parser')

    error_div = soup.find("div", class_="error-inner")
    if error_div:
      status = "에러"
      results.append({'URL':url, '상태':status})
    else:
      status = "Ok"
      results.append({'URL':url, '상태':status})

  except Exception as e:
    status = f"RequestError:{e}"
  if idx % 100 ==0:
    print(f'{idx}/{len(urls)} processed')


    # time.sleep(0.2)
```
