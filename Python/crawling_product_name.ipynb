```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
base_url = "https://lgq.kr/GS_"

all_links = ["20150356465193", "20152048348545", "20150688736699"]


data = []

for link in all_links:
  url = base_url + link
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  title = soup.find('h2', class_='head-title')

  if title:
    text = title.get_text(strip=True)
    result = text.split("[")[1].split("]")[0]
  else:
    result = "없음"

  data.append({link,result})
data

df = pd.DataFrame(data)

df
```
