```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
all_links =["20153015582211",
"20153014256312",
"20150370401717",
"20150806991006",
"20153015592242",
"20153015500482",
"20150791262992",
"20153004592447",
"20153030868197",
"20153014649039"]
```

```python
base_url = "https://lgq.kr/GS_"



data = []

for link in all_links:
  url = base_url + link
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  title = soup.find('h2', class_='head-title')

  if title:
    text = title.get_text(strip=True)
    if "[" in text:
      result = text.split("[")[1].split("]")[0]
    else:
      result = "기호없음"
  else:
    result = "에러"

  data.append([link,result])


df = pd.DataFrame(data)
```

```python
df.to_csv('data.csv', index=False, encoding='utf-8-sig')
```
