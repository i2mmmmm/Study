```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display
```

```python
df = pd.read_excel('base1.xlsx',sheet_name=None, index_col=0)
```

```python
test1 = df['Sheet1']
test2 = df['Sheet2']
test3 = df['Sheet3']
test4 = df['Sheet4']
```

```python
# 모든 시트를 하나의 리스트로 관리
all = [test1, test2, test3, test4]

# 행, 열에 따른 합계 구하기
for i, df in enumerate(all, start=1):
    # 각 열의 합계 계산하여 'sum' 행 추가
    df.loc['sum_col'] = df.sum()
    # 각 행의 합계 계산하여 'sum' 열 추가
    df['sum_row'] = df.iloc[:, 1:].sum(axis=1)
```

```python
#  합계 합친 dataframe 만들기
result = pd.concat([test1.iloc[-1:,:-1], test2.iloc[-1:,:-1], test3.iloc[-1:,:-1], test4.iloc[-1:,:-1]], axis=0)
```

```python
df2 = pd.read_excel('base1.xlsx',sheet_name=None, index_col=0)

test_1 = df2['Sheet1']
test_2 = df2['Sheet2']
test_3 = df2['Sheet3']
test_4 = df2['Sheet4']
```

```python
dfs = {'IB': test_1, 'OB': test_2, 'clame': test_3, 'request': test_4}

# 사용자 입력 받기
data_choice = input("어떤 데이터를 보고 싶으세요? (IB, OB, clame, request): ").strip()
week_choice = input("몇 주차를 보고 싶으세요?(숫자만 입력해주세요): ").strip()

# 유효성 검사
if data_choice not in dfs:
    print("잘못된 데이터 선택입니다. IB, OB, clame, request 중 하나를 입력해주세요.")
elif not week_choice.isdigit() or int(week_choice) < 1 or int(week_choice) > 10:
    print("잘못된 숫자입니다. 1에서 10 사이의 숫자를 입력하세요.")
else:
    week_num = int(week_choice)
    df = dfs[data_choice]
    week_label = f'{week_num}주차'

    if week_label not in df.index:
        print(f"'{week_label}' 주차는 존재하지 않습니다.")
    else:
        row = df.loc[week_label]
        top3 = row.nlargest(3)
        print(f"\n{data_choice}의 {week_label} 상위 3위:")
        print(top3)
```

```python
# 위젯 생성
data_choice_widget = widgets.Dropdown(
    options=['IB', 'OB', 'clame', 'request'],
    value='IB',
    description='DataFrame:',
)

week_choice_widget = widgets.IntSlider(
    value=1,
    min=1,
    max=10,
    step=1,
    description='주차:',
    style={'description_width': 'initial'}
)

output = widgets.Output()

def update_result(change):
    with output:
        output.clear_output()
        data_choice = data_choice_widget.value
        week_num = week_choice_widget.value
        df = dfs[data_choice]
        week_label = f'{week_num}주차'

        if week_label not in df.index:
            print(f"'{week_label}' 주차는 존재하지 않습니다.")
        else:
            row = df.loc[week_label]
            top3 = row.nlargest(3)
            print(f"\n{data_choice}의 {week_label} 상위 3위:")
            print(top3)

# 위젯과 출력 연결
data_choice_widget.observe(update_result, names='value')
week_choice_widget.observe(update_result, names='value')

# 위젯과 결과 출력
display(data_choice_widget)
display(week_choice_widget)
display(output)
```
